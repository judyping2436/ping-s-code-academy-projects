Seaborn Visualizing World Cup Feb2021 Ping

Exploring data from the Fifa World Cup from 1930-2014 to analyze trends and discover insights about the world’s game, fútbol!


```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Load Data from csv
df = pd.read_csv('WorldCupMatches.csv', delimiter = ',')
goal = pd.read_csv('goals.csv', delimiter = ',')
df.head(2)
#goal.head(2)
```

```python
goal.head(2)
```

```python
# Create a 'Total Goals' column
df['Total Goals'] = df['Home Team Goals'] + df['Away Team Goals']
df.head(2)
```

```python
# Create a bar chart

plt.figure(figsize=(12, 7)) 
sns.set_style('whitegrid')
sns.set_context('poster', font_scale = .8)
ax = sns.barplot(data = df, x = 'Year', y = 'Total Goals')
ax.set_title("Average Number of Goals Scored In World Cup Matches 1930 - 2014")
ax.set_ylabel('Average Goals Per Match')
ax.set_xticklabels(df['Year'].unique(), rotation = 30)
plt.show()
```

```python
# Create a box chart from goals.csv
plt.clf()
plt.figure(figsize = (12, 7))
sns.set_style('whitegrid')
sns.set_context('notebook', font_scale = 1.25)
ax = sns.boxplot(data = goal, x = "year", y = 'goals', palette='Spectral')
ax.set_title("Average Number of Goals Scored in World Cup Matches 1930 - 2014")
ax.set_xticklabels(goal['year'].unique(), rotation = 30)
plt.show()
plt.clf()
```
